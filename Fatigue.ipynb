```python
import numpy as np
import matplotlib.pyplot as plt
Syt=330
Se=250
Sut=450
plt.figure(figsize=(7,7))
def soderberg(syt,se):
    plt.plot([syt,0],[0,se],label="soderberg line",color='blue')
    plt.legend()
    plt.xlim(0,Sut+50)
    plt.xlabel(r'$\sigma_m$')
    plt.ylabel(r'$\sigma_a$')
    plt.ylim(0,se+50)
    return
def goodman(sut,se):
    plt.plot([sut,0],[0,se],label="goodman line",color='red')
    plt.legend()
    return
def gerber(Se,Sut):
    Sm=np.linspace(0,Sut,100)
    Sa=(1-(Sm/Sut)**2)*Se
    plt.plot(Sm,Sa,label="Gerber line",color='black')
    plt.legend()
    return
soderberg(Syt,Se)
goodman(Sut,Se)
gerber(Se,Sut)
#checking the criteria for different mean and amplitude stresses
S_check=[100,200,178,90]
s_checkampli=[90,220,258,70]
plt.scatter(S_check,s_checkampli)
             
```
