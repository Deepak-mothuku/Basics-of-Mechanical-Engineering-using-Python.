```python
import numpy as np
import matplotlib.pyplot as plt
syt=200; #Yield strength
s1=150; #stress in dir1
s2=80; #stress in direction2
sv=np.sqrt(s1**2-s1*s2+s2**2) #vonmises
print(f"the vonmises stress for given stress is{sv}")
a=2**0.5*syt #semimajoraxisofellipse
b=(2/3)**0.5*syt #semiminoraxisofellipse
theta=np.linspace(0,360*np.pi/180,720);
x=a*np.cos(theta); #originalpoint
y=b*np.sin(theta); #originalpoint
theta1=np.pi/4;
x_new=x*np.cos(theta1)-y*np.sin(theta1) #transformed
y_new=x*np.sin(theta1)+y*np.cos(theta1)
xp=[syt,syt,0,-syt,-syt,0,syt] #maximum shear stresstheoryxcoord
yp=[0,syt,syt,0,-syt,-syt,0] #ycoord
plt.figure(figsize=(7.5,5));
plt.plot(x_new,y_new,label="vonmises_region");
plt.plot(xp,yp,label="max_shear_stress",linestyle="dashed");
plt.legend()
xc=[200,160,80,60] #checking the stresses in x,y directions are safe
yc=[-80,210,50,40]
plt.scatter(xc,yc)
plt.show()




```
